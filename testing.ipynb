```python
import numpy as np
from nnfs.datasets import spiral_data
from moonet import MooNet as mn, Relu, SoftMax
```

```python
# X is the input data, y is the labeled output data
X,y = spiral_data(samples=100, classes=3)

# defining layer 1, with 2 input features and 3 neurons
dense1 = mn(2,3)
# defining layer 2, with 3 input(output of layer 1) and 3 neurons
dense2 = mn(3,3)
# activation functions
activation1 = Relu()
activation2 = SoftMax()

# passing data through layer 1
dense1.forward_pass(X)
# applying relu activation on layer 1
activation1.forward_pass(dense1.output)
# passing the output of layer 1 to layer 2
dense2.forward_pass(activation1.output)
# this being the final layer, we apply softmax activation
activation2.forward_pass(activation1.output)


print(activation2.output)
```
